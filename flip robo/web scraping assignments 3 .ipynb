 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then
scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Initialize Chrome WebDriver 
driver = webdriver.Chrome()

# Initialize a list to store the scraped data
products_data = []

# Keywords to search for products
keywords = ['Guitar']

# Function to search for products and scrape details
def search_and_scrape_products(search_query, num_pages=3):
    for page in range(1, num_pages + 1):
        # Open Google Shopping
        driver.get("https://www.google.com/shopping")
        
        # Find the search input field and enter the search query
        search_input = driver.find_element(By.NAME, "q")
        search_input.clear()
        search_input.send_keys(search_query)
        
        # Press Enter key to perform the search
        search_input.send_keys(Keys.RETURN)
        
        # Wait for the search results to load
        time.sleep(3)
        
        # Scroll down the page to load more products 
        for _ in range(page):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)
        
        # Find all product entries on the page
        product_entries = driver.find_elements(By.CSS_SELECTOR, "div.sh-dlr__list-result")
        
        # Loop through each product entry and extract details
        for product_entry in product_entries:
            brand_name = product_entry.find_element(By.CSS_SELECTOR, "div.dsh-details span.sxp-dsh-title").text
            product_name = product_entry.find_element(By.CSS_SELECTOR, "div.dsh-details span.sxp-dsh-title").text
            price = product_entry.find_element(By.CSS_SELECTOR, "div.dsh-price span").text
            return_exchange = product_entry.find_element(By.CSS_SELECTOR, "div.dsh-returns span").text
            expected_delivery = product_entry.find_element(By.CSS_SELECTOR, "div.dsh-expected span").text
            availability = product_entry.find_element(By.CSS_SELECTOR, "div.dsh-availability span").text
            product_url = product_entry.find_element(By.CSS_SELECTOR, "a").get_attribute("href")
            
            product_data = {
                "Brand Name": brand_name,
                "Name of the Product": product_name,
                "Price": price,
                "Return/Exchange": return_exchange,
                "Expected Delivery": expected_delivery,
                "Availability": availability,
                "Product URL": product_url,
            }
            
            products_data.append(product_data)

# Scrape products for each keyword
for keyword in keywords:
    search_and_scrape_products(keyword, num_pages=3)

# Close the browser
driver.quit()

# Create a DataFrame to store the scraped data
df = pd.DataFrame(products_data)

# Save the data to a CSV file
df.to_csv("products.csv", index=False)

# Print the DataFrame
print(df)
```

 3. Write a python program to access the search bar and search button on images.google.com and scrape 10
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’. 

```python
import time
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import pandas as pd

# Initialize Chrome WebDriver 
driver = webdriver.Chrome()

# Open Google Images
driver.get("https://www.google.com/imghp")

# Function to search for images and scrape them
def search_and_scrape_images(search_query, num_images=10):
    # Find the search input field and enter the search query
    search_input = driver.find_element(By.NAME, "q")
    search_input.clear()
    search_input.send_keys(search_query)
    
    # Press Enter key to perform the search
    search_input.send_keys(Keys.RETURN)
    
    # Wait for the search results to load
    time.sleep(3)
    
    # Scroll down the page to load more images 
    for _ in range(num_images // 10):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
    
    # Find and scrape the image URLs
    image_elements = driver.find_elements(By.CSS_SELECTOR, ".rg_i")
    image_urls = []
    
    for i, image_element in enumerate(image_elements[:num_images], start=1):
        image_url = image_element.get_attribute("src")
        if image_url:
            image_urls.append(image_url)
    
    # Create a directory to save the images
    os.makedirs(search_query, exist_ok=True)
    
    # Download and save the images
    for i, image_url in enumerate(image_urls, start=1):
        image_path = os.path.join(search_query, f"{search_query}_{i}.jpg")
        try:
            driver.get(image_url)
            with open(image_path, "wb") as file:
                file.write(driver.page_source)
        except Exception as e:
            print(f"Failed to download image {i} for '{search_query}': {str(e)}")
    
    return image_urls

# Keywords to search for images
keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']

# Scrape images for each keyword
for keyword in keywords:
    print(f"Scraping images for '{keyword}'...")
    search_and_scrape_images(keyword, num_images=10)

# Close the browser
driver.quit()
```

4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand
Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the
details is missing then replace it by “- “. Save your results in a dataframe and CSV. 


```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open Flipkart
driver.get("https://www.flipkart.com/")

# Function to search for a smartphone and scrape details
def search_and_scrape_smartphones(search_query):
    # Find the search input field and enter the search query
    search_input = driver.find_element(By.NAME, "q")
    search_input.clear()
    search_input.send_keys(search_query)
    
    # Find and click the search button
    search_button = driver.find_element(By.CLASS_NAME, "L0Z3Pu")
    search_button.click()
    
    # Wait for the search results to load
    time.sleep(3)
    
    # Find all smartphone results on the first page
    smartphone_results = driver.find_elements(By.CSS_SELECTOR, "div._1AtVbE")
    
    # Initialize empty lists to store smartphone details
    brand_names = []
    smartphone_names = []
    colors = []
    rams = []
    storages = []
    primary_cameras = []
    secondary_cameras = []
    display_sizes = []
    battery_capacities = []
    prices = []
    product_urls = []
    
    # Loop through each smartphone result and extract details
    for result in smartphone_results:
        # Extract brand and name
        brand_name, smartphone_name = result.find_element(By.CSS_SELECTOR, "div._4rR01T").text.split(' ')
        
        # Extract other details
        details = result.find_elements(By.CSS_SELECTOR, "ul._1xgFaf li")
        color = details[0].text if len(details) > 0 else "-"
        ram = details[1].text if len(details) > 1 else "-"
        storage = details[2].text if len(details) > 2 else "-"
        primary_camera = details[3].text if len(details) > 3 else "-"
        secondary_camera = details[4].text if len(details) > 4 else "-"
        display_size = details[5].text if len(details) > 5 else "-"
        battery_capacity = details[6].text if len(details) > 6 else "-"
        price = result.find_element(By.CSS_SELECTOR, "div._30jeq3").text
        product_url = result.find_element(By.CSS_SELECTOR, "a._1fQZEK").get_attribute("href")
        
        # Append details to lists
        brand_names.append(brand_name)
        smartphone_names.append(smartphone_name)
        colors.append(color)
        rams.append(ram)
        storages.append(storage)
        primary_cameras.append(primary_camera)
        secondary_cameras.append(secondary_camera)
        display_sizes.append(display_size)
        battery_capacities.append(battery_capacity)
        prices.append(price)
        product_urls.append(product_url)
    
    # Create a DataFrame to store the scraped data
    df = pd.DataFrame({
        "Brand Name": brand_names,
        "Smartphone Name": smartphone_names,
        "Colour": colors,
        "RAM": rams,
        "Storage(ROM)": storages,
        "Primary Camera": primary_cameras,
        "Secondary Camera": secondary_cameras,
        "Display Size": display_sizes,
        "Battery Capacity": battery_capacities,
        "Price": prices,
        "Product URL": product_urls,
    })
    
    # Save the data to a CSV file
    df.to_csv("smartphones.csv", index=False)
    
    return df

# Input the smartphone you want to search for
search_query = "Oneplus Nord"  # Replace with the desired smartphone

# Search for the smartphone and scrape details
smartphone_df = search_and_scrape_smartphones(search_query)

# Close the browser
driver.quit()

# Print the DataFrame
print(smartphone_df)
```

 6. Write a program to scrap all the available details of best gaming laptops from digit.in. 

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import pandas as pd

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Open the Digit.in page for gaming laptops
driver.get("https://www.digit.in/top-products/best-gaming-laptops-40.html")

# Function to scroll to the end of the page to load all laptop details
def scroll_to_end():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Scroll down repeatedly until all laptops are loaded 
while True:
    scroll_to_end()
    try:
        # Look for the "Load More" button and click it if it exists
        load_more_button = driver.find_element(By.CSS_SELECTOR, ".page .more-btn")
        load_more_button.click()
        time.sleep(2)
    except Exception as e:
        break  # No more "Load More" button found, exit the loop

# Find all laptop entries
laptop_entries = driver.find_elements(By.CSS_SELECTOR, ".TopNumbeHeading")

# Initialize empty list to store laptop data
laptops_data = []

# Loop through each laptop entry and extract data
for laptop_entry in laptop_entries:
    name = laptop_entry.text
    specs_element = laptop_entry.find_element(By.XPATH, "./following-sibling::p")
    specs = specs_element.text

    laptop_data = {
        "Name": name,
        "Specifications": specs,
    }

    laptops_data.append(laptop_data)

# Close the browser
driver.quit()

# Print or save the scraped data as needed
for i, laptop in enumerate(laptops_data, start=1):
    print(f"Laptop {i}:")
    print(f"Name: {laptop['Name']}")
    print(f"Specifications: {laptop['Specifications']}")
    print("\n")
```

7. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped:
“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import pandas as pd

# Initialize Chrome WebDriver 
driver = webdriver.Chrome()

# Open Forbes Billionaires page
driver.get("https://www.forbes.com/billionaires/")

# Function to scroll to the end of the page to load all billionaire entries
def scroll_to_end():
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(2)

# Scroll down repeatedly until all billionaires are loaded 
while True:
    scroll_to_end()
    try:
        # Look for the "Load More" button and click it if it exists
        load_more_button = driver.find_element(By.XPATH, "//button[contains(text(), 'Load More')]")
        load_more_button.click()
        time.sleep(2)
    except Exception as e:
        break  # No more "Load More" button found, exit the loop

# Find all billionaire entries
billionaires = driver.find_elements(By.CSS_SELECTOR, ".billionaires-list .table-row")

# Initialize empty list to store billionaire data
billionaires_data = []

# Loop through each billionaire entry and extract data
for billionaire in billionaires:
    rank = billionaire.find_element(By.CSS_SELECTOR, ".rank").text
    name = billionaire.find_element(By.CSS_SELECTOR, ".name").text
    net_worth = billionaire.find_element(By.CSS_SELECTOR, ".net-worth").text
    age = billionaire.find_element(By.CSS_SELECTOR, ".age").text
    citizenship = billionaire.find_element(By.CSS_SELECTOR, ".country-of-citizenship").text
    source = billionaire.find_element(By.CSS_SELECTOR, ".source").text
    industry = billionaire.find_element(By.CSS_SELECTOR, ".category").text

    billionaire_data = {
        "Rank": rank,
        "Name": name,
        "Net Worth": net_worth,
        "Age": age,
        "Citizenship": citizenship,
        "Source": source,
        "Industry": industry,
    }

    billionaires_data.append(billionaire_data)

# Close the browser
driver.quit()

# Print or save the scraped data as needed
for i, billionaire in enumerate(billionaires_data, start=1):
    print(f"Billionaire {i}:")
    for key, value in billionaire.items():
        print(f"{key}: {value}")
    print("\n")
```

8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted
from any YouTube Video. 

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import pandas as pd

# URL of the YouTube video you want to scrape comments from
video_url = "https://www.youtube.com/watch?v=YOUR_VIDEO_ID"

# Initialize Chrome WebDriver 
driver = webdriver.Chrome()

# Open the YouTube video page
driver.get(video_url)

# Scroll to load more comments (you may adjust the number of scrolls based on your needs)
while True:
    try:
        # Locate the "Show more" button and click it to load more comments
        show_more_button = driver.find_element(By.XPATH, "//yt-formatted-string[contains(text(), 'Show more')]")
        show_more_button.click()
        time.sleep(2)  # Wait for comments to load
    except Exception as e:
        break  # If no more comments to load, exit the loop

# Initialize empty lists to store comments, upvotes, and timestamps
comments = []
upvotes = []
timestamps = []

# Extract comments, upvotes, and timestamps
comment_elements = driver.find_elements(By.CSS_SELECTOR, "#content-text")
upvote_elements = driver.find_elements(By.CSS_SELECTOR, "#vote-count-middle")
timestamp_elements = driver.find_elements(By.CSS_SELECTOR, "#header-author > yt-formatted-string > a > #published-time-text")

for comment_element, upvote_element, timestamp_element in zip(comment_elements, upvote_elements, timestamp_elements):
    comments.append(comment_element.text)
    upvotes.append(upvote_element.text)
    timestamps.append(timestamp_element.get_attribute("aria-label"))

# Close the browser
driver.quit()

# Print or save the scraped data as needed
for i in range(len(comments)):
    print(f"Comment {i + 1}:")
    print(f"Comment: {comments[i]}")
    print(f"Upvotes: {upvotes[i]}")
    print(f"Timestamp: {timestamps[i]}")
    print("\n")
```

9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall
reviews, privates from price, dorms from price, facilities and property description. 

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import pandas as pd

# Initialize Chrome WebDriver 
driver = webdriver.Chrome()

# Define the URL for London hostels on Hostelworld
url = "https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3"

# Open the URL in the browser
driver.get(url)

# Function to scroll to the bottom of the page to load more hostels
def scroll_to_bottom():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Scroll to load more hostels 
for _ in range(3):
    scroll_to_bottom()

# Find all hostel elements on the page
hostel_elements = driver.find_elements(By.CSS_SELECTOR, ".property-card")

# Initialize an empty list to store hostel data
hostels_data = []

# Loop through each hostel element and extract data
for hostel_element in hostel_elements:
    hostel_data = {}
    
    # Extract hostel name
    hostel_data['Name'] = hostel_element.find_element(By.CSS_SELECTOR, ".property-card-body .title-2").text
    
    # Extract distance from city centre
    hostel_data['Distance'] = hostel_element.find_element(By.CSS_SELECTOR, ".property-card-body .subtitle-2").text
    
    # Extract ratings and total reviews
    ratings = hostel_element.find_element(By.CSS_SELECTOR, ".rating").text.split()
    hostel_data['Ratings'] = ratings[0]
    hostel_data['Total Reviews'] = ratings[1]
    
    # Extract overall reviews
    hostel_data['Overall Reviews'] = hostel_element.find_element(By.CSS_SELECTOR, ".reviews .rating").text
    
    # Extract prices for privates and dorms
    prices = hostel_element.find_elements(By.CSS_SELECTOR, ".prices .price")
    hostel_data['Privates Price'] = prices[0].text
    hostel_data['Dorms Price'] = prices[1].text
    
    # Extract facilities
    facilities = [facility.text for facility in hostel_element.find_elements(By.CSS_SELECTOR, ".facilities .label")]
    hostel_data['Facilities'] = ', '.join(facilities)
    
    # Extract property description
    hostel_data['Property Description'] = hostel_element.find_element(By.CSS_SELECTOR, ".description").text
    
    # Append the hostel data to the list
    hostels_data.append(hostel_data)

# Close the browser
driver.quit()

# Print the scraped data
for i, hostel in enumerate(hostels_data, start=1):
    print(f"Hostel {i}:")
    for key, value in hostel.items():
        print(f"{key}: {value}")
    print("\n")

# 
```
