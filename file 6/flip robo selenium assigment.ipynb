```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize Chrome WebDriver
chrome_service = ChromeService("path_to_chromedriver")
chrome_options = ChromeOptions()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (no GUI)
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

# Initialize variables
base_url = "https://www.flipkart.com"
search_query = "sunglasses"
max_items = 100  # Maximum items to scrape

# Initialize lists to store data
brands = []
descriptions = []
prices = []

# Step 1: Go to Flipkart webpage
driver.get(base_url)

# Step 2: Enter "sunglasses" in the search field and click the search icon
search_field = driver.find_element(By.NAME, "q")
search_field.clear()
search_field.send_keys(search_query)
search_field.send_keys(Keys.RETURN)

# Step 3, 4, 5: Loop until you have data for 100 sunglasses
page_number = 1
while len(brands) < max_items:
    # Wait for the page to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "_1AtVbE")))

    # Find and extract product details
    product_containers = driver.find_elements(By.CLASS_NAME, "_1AtVbE")

    for container in product_containers:
        # Extract brand
        brand = container.find_element(By.CLASS_NAME, "_2WkVRV").text

        # Extract product description
        description = container.find_element(By.CLASS_NAME, "IRpwTa").text

        # Extract price
        price = container.find_element(By.CLASS_NAME, "_30jeq3").text

        # Append data to lists
        brands.append(brand)
        descriptions.append(description)
        prices.append(price)

    # Move to the next page if available
    try:
        next_button = driver.find_element(By.LINK_TEXT, "Next")
        next_button.click()
    except NoSuchElementException:
        break

    # Increment the page number
    page_number += 1

# Step 6: Create a DataFrame with the scraped data
sunglasses_data = pd.DataFrame({
    "Brand": brands[:max_items],
    "Product Description": descriptions[:max_items],
    "Price": prices[:max_items]
})

# Display the DataFrame
print(sunglasses_data)

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize Chrome WebDriver
chrome_service = ChromeService("path_to_chromedriver")
chrome_options = ChromeOptions()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (no GUI)
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

# Initialize variables
base_url = "https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART"
max_reviews = 100  # Maximum reviews to scrape

# Initialize lists to store data
ratings = []
review_summaries = []
full_reviews = []

# Go to the provided Flipkart link
driver.get(base_url)

# Wait for the page to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "_27M-vq")))

# Loop until you have data for 100 reviews
while len(ratings) < max_reviews:
    # Find and extract review details
    review_containers = driver.find_elements(By.CLASS_NAME, "_27M-vq")

    for container in review_containers:
        # Extract rating
        rating = container.find_element(By.CLASS_NAME, "_3LWZlK").text

        # Extract review summary
        summary = container.find_element(By.CLASS_NAME, "col").text

        # Click on "Read more" to see full review
        read_more_button = container.find_element(By.XPATH, ".//span[contains(text(),'Read more')]")
        driver.execute_script("arguments[0].click();", read_more_button)

        # Extract full review
        full_review = container.find_element(By.CLASS_NAME, "t-ZTKy").text

        # Append data to lists
        ratings.append(rating)
        review_summaries.append(summary)
        full_reviews.append(full_review)

        # Check if we have enough reviews
        if len(ratings) == max_reviews:
            break

    # Navigate to the next page of reviews
    try:
        next_button = driver.find_element(By.CLASS_NAME, "_1LKTO3")
        driver.execute_script("arguments[0].click();", next_button)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "_27M-vq")))
    except NoSuchElementException:
        break

# Create a DataFrame with the scraped data
reviews_data = pd.DataFrame({
    "Rating": ratings,
    "Review Summary": review_summaries,
    "Full Review": full_reviews
})

# Display the DataFrame
print(reviews_data)

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize Chrome WebDriver
chrome_service = ChromeService("path_to_chromedriver")
chrome_options = ChromeOptions()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (no GUI)
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

# Initialize variables
base_url = "https://www.amazon.in"
search_query = "Laptop"
max_items = 10  # Maximum items to scrape

# Initialize lists to store data
titles = []
ratings = []
prices = []

# Step 1: Go to Amazon.in webpage
driver.get(base_url)

# Step 2: Enter "Laptop" in the search field and click the search icon
search_field = driver.find_element(By.ID, "twotabsearchtextbox")
search_field.clear()
search_field.send_keys(search_query)
search_field.send_keys(Keys.RETURN)

# Step 3: Set CPU Type filter to "Intel Core i7"
cpu_filter = driver.find_element(By.XPATH, "//li[contains(., 'Intel Core i7')]")

actions = ActionChains(driver)
actions.move_to_element(cpu_filter).click().perform()

# Loop until you have data for 10 laptops
while len(titles) < max_items:
    # Wait for the page to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "s-image")))

    # Find and extract product details
    product_containers = driver.find_elements(By.XPATH, "//div[@data-component-type='s-search-result']")

    for container in product_containers:
        # Extract title
        title = container.find_element(By.XPATH, ".//h2/a").text

        # Extract rating
        try:
            rating = container.find_element(By.CLASS_NAME, "a-popover-trigger").get_attribute("aria-label")
        except:
            rating = "N/A"

        # Extract price
        try:
            price = container.find_element(By.XPATH, ".//span[@class='a-price-whole']").text
        except:
            price = "N/A"

        # Append data to lists
        titles.append(title)
        ratings.append(rating)
        prices.append(price)

    # Check if we have enough data
    if len(titles) >= max_items:
        break

    # Move to the next page if available
    try:
        next_button = driver.find_element(By.CLASS_NAME, "s-pagination-next")
        next_button.click()
    except NoSuchElementException:
        break

# Create a DataFrame with the scraped data
laptop_data = pd.DataFrame({
    "Title": titles[:max_items],
    "Ratings": ratings[:max_items],
    "Price": prices[:max_items]
})

# Display the DataFrame
print(laptop_data)

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize Chrome WebDriver
chrome_service = ChromeService("path_to_chromedriver")
chrome_options = ChromeOptions()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (no GUI)
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

# Initialize the URL
url = "https://www.azquotes.com/"

# Step 1: Get the webpage
driver.get(url)

# Step 2: Click on "Top Quotes"
top_quotes_link = driver.find_element(By.LINK_TEXT, "Top Quotes")
top_quotes_link.click()

# Wait for the page to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CLASS_NAME, "wrap-block")))

# Initialize lists to store data
quotes = []
authors = []
quote_types = []

# Loop through the pages to scrape data for the top 1000 quotes
page_number = 1
quotes_to_scrape = 1000

while len(quotes) < quotes_to_scrape:
    # Find all the quote elements on the page
    quote_elements = driver.find_elements(By.CLASS_NAME, "wrap-block")

    for quote_element in quote_elements:
        # Extract quote text
        quote = quote_element.find_element(By.CLASS_NAME, "title").text

        # Extract author name
        author = quote_element.find_element(By.CLASS_NAME, "author").text

        # Extract type of quote
        quote_type = quote_element.find_element(By.CLASS_NAME, "kw-box").text

        # Append data to lists
        quotes.append(quote)
        authors.append(author)
        quote_types.append(quote_type)

        # Check if we have scraped the desired number of quotes
        if len(quotes) >= quotes_to_scrape:
            break

    # Navigate to the next page if available
    try:
        next_button = driver.find_element(By.LINK_TEXT, "Next ")
        next_button.click()
        page_number += 1
    except NoSuchElementException:
        break

# Create a DataFrame with the scraped data
quotes_data = pd.DataFrame({
    "Quote": quotes,
    "Author": authors,
    "Type Of Quote": quote_types
})

# Display the first few rows of the DataFrame
print(quotes_data.head())

# Close the WebDriver
driver.quit()
```
